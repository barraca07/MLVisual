# Visualise Classification Algorithms Using 2D Synthetic Dataset

```python
from utils import visualise_synthetic

from LogisticRegression import LogisticRegression
from NearestNeighbours import NearestNeighbours
from NaiveBayes import NaiveBayes
from DecisionTree import DecisionTree
from RandomForest import RandomForest
from AdaBoost import AdaBoost
from SupportVectorMachine import SupportVectorMachine
from NeuralNetwork import NeuralNetwork
```

## Logistic Regression

```python
visualise_synthetic(LogisticRegression(), "Logistic Regression")
```

## K-Nearest Neighbours

```python
visualise_synthetic(NearestNeighbours(k=9), "KNN")
```

## Naive Bayes

```python
visualise_synthetic(NaiveBayes(), "Naive Bayes")
```

## Decision Tree

```python
visualise_synthetic(DecisionTree(max_depth=2), "Decision Tree")
```

## Random Forest

```python
random_forest_model = RandomForest(num_trees=30, max_depth=2, samples_fraction=0.5, features_fraction=1)

visualise_synthetic(random_forest_model, "Random Forest")
```

## AdaBoost

```python
visualise_synthetic(AdaBoost(num_learners=15), "AdaBoost")
```

## Support Vector Machine

```python
visualise_synthetic(SupportVectorMachine(C=0.01), "Support Vector Machine")
```

## Neural Network
